```python
df = data.iloc[:,2:]
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']]) #passed lists with order
#created obj
```

```python
oe.fit(x_train)#train ke upr fit krte ho aur train aur test dono ke upr transform krte ho
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```
